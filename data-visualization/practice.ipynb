```python
from matplotlib import pylab
import numpy as np
```

```python
X = np.linspace(0, 10, 25)
y = X * X + 2
print(np.array([X, y]).reshape(25, 2))
```

```python
pylab.plot(X, y, 'r')
```

```python
pylab.subplot(1, 2, 1)
pylab.plot(X, y, 'r--')
pylab.subplot(1, 2, 2)
pylab.plot(y, X, 'g--')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axes.plot(X, y, 'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
[a.plot(X, y, 'r') for a in axes]
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(X, y, 'r')
axes2.plot(X, y, 'g')
```

```python
fig, axes = plt.subplots(1, 2, figsize =(10, 5))

axes[0].plot(X, X**2, X, X**3, lw=1)
axes[0].grid(True)

axes[1].plot(X, X**2, X, X**3, lw=1)
axes[1].grid(True)
```

```python
n = np.array([1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize =(16, 5))
axes[0].set_title("Scatter")
axes[0].scatter(n, n**2, lw=1)

axes[1].step(n, n**2, lw=1)
axes[1].set_title("Step")

axes[2].hist(n, n**2, lw=1)
axes[2].set_title("Histogram")

axes[3].fill_between(n, n**2, lw=1)
axes[3].set_title('Fill Between')
```

```python
fig =  plt.figure(figsize=(6,6))
axes = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 2*np.pi, 100)
axes.plot(t, np.sin(t))
```
